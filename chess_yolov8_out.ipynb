```
import torch
from PIL import Image
from ultralytics import YOLO

# Path to the model weights (your pre-trained model)
model_path = "/content/drive/MyDrive/yolov8/best (3).pt"  # Adjust according to your path
model = YOLO(model_path)

# Path to test image
image_path = "/content/drive/MyDrive/yolov8/test/images/IMG_0595_JPG.rf.b75a8d0ab132cb206de413213ba8c721.jpg"  # Adjust the path to your test image
img = Image.open(image_path)

# Perform inference
results = model(img)

# Show the results (access the first element of the results list)
results[0].show()  # Display the annotated image

# Extract predictions
boxes = results[0].boxes.xywh  # Bounding boxes (in xywh format)
labels = results[0].boxes.cls  # Predicted class IDs
scores = results[0].boxes.conf  # Confidence scores

# Mapping of class labels to class names
class_names = [
    "Black-Bishop", "Black-King", "Black-Knight", "Black-Pawn", "Black-Queen",
    "Black-Rook", "White-Bishop", "White-King", "White-Knight", "White-Pawn",
    "White-Queen", "White-Rook"
]

# Loop through the detected boxes and display the results
for i, box in enumerate(boxes):
    class_id = int(labels[i].item())  # Get the predicted class ID
    score = scores[i].item()  # Confidence score
    print(f"Detection {i+1}:")
    print(f"Class: {class_names[class_id]}")
    print(f"Bounding Box (xywh): {box}")
    print(f"Confidence Score: {score:.4f}")
    print()
```
